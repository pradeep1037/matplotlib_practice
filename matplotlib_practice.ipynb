```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
medical_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Medicine_Distributed': [1200, 1500, 1800, 1700, 2100, 2500],
    'Hospital_Orders': [30, 45, 50, 48, 60, 75],
    'Delivery_Time': [5, 4, 6, 5, 3, 2],
    'Revenue': [50000, 65000, 72000, 70000, 85000, 95000]
}

df = pd.DataFrame(medical_data)

print(df)
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    df['Month'],
    df['Medicine_Distributed'],
    marker='o'
)

plt.title('Medicine Distribution Over Months')
plt.xlabel('Month')
plt.ylabel('Medicine Distributed')
plt.grid(True)

plt.savefig('medicine_distribution.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    df['Month'],
    df['Revenue'],
    marker='o'
)

plt.title('Monthly Revenue Growth')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)

plt.savefig('revenue_growth.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    df['Month'],
    df['Medicine_Distributed'],
    marker='o',
    label='Medicine Distributed'
)

plt.plot(
    df['Month'],
    df['Hospital_Orders'],
    marker='s',
    label='Hospital Orders'
)

plt.title('Medicine Distributed vs Hospital Orders')
plt.xlabel('Month')
plt.ylabel('Values')

plt.legend()
plt.grid(True)

plt.savefig('multiple_line_chart.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    df['Month'],
    df['Hospital_Orders']
)

plt.title('Hospital Orders by Month')
plt.xlabel('Month')
plt.ylabel('Orders')

plt.savefig('hospital_orders_bar.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.barh(
    df['Month'],
    df['Revenue']
)

plt.title('Revenue by Month')
plt.xlabel('Revenue')
plt.ylabel('Month')

plt.savefig('revenue_horizontal_bar.png')

plt.show()
```

```python
x = np.arange(len(df['Month']))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(
    x - width/2,
    df['Medicine_Distributed'],
    width,
    label='Medicine Distributed'
)

plt.bar(
    x + width/2,
    df['Hospital_Orders'],
    width,
    label='Hospital Orders'
)

plt.xticks(x, df['Month'])

plt.title('Medicine Distribution vs Hospital Orders')
plt.xlabel('Month')
plt.ylabel('Values')

plt.legend()

plt.savefig('grouped_bar_chart.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Hospital_Orders'],
    df['Revenue']
)

plt.title('Hospital Orders vs Revenue')
plt.xlabel('Hospital Orders')
plt.ylabel('Revenue')

plt.savefig('orders_vs_revenue.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Delivery_Time'],
    df['Revenue']
)

plt.title('Delivery Time vs Revenue')
plt.xlabel('Delivery Time')
plt.ylabel('Revenue')

plt.savefig('delivery_vs_revenue.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Medicine_Distributed'],
    df['Hospital_Orders']
)

plt.title('Medicine Distributed vs Hospital Orders')
plt.xlabel('Medicine Distributed')
plt.ylabel('Hospital Orders')

plt.savefig('medicine_vs_orders.png')

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.pie(
    df['Medicine_Distributed'],
    labels=df['Month'],
    autopct='%1.1f%%'
)

plt.title('Medicine Distribution Percentage')

plt.savefig('medicine_distribution_pie.png')

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.pie(
    df['Revenue'],
    labels=df['Month'],
    autopct='%1.1f%%'
)

plt.title('Revenue Contribution by Month')

plt.savefig('revenue_contribution_pie.png')

plt.show()
```

```python
explode = [0, 0, 0, 0, 0, 0.2]

plt.figure(figsize=(8,6))

plt.pie(
    df['Revenue'],
    labels=df['Month'],
    autopct='%1.1f%%',
    explode=explode
)

plt.title('Highest Revenue Month Highlighted')

plt.savefig('highest_revenue_pie.png')

plt.show()
```
